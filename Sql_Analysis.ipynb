# Show All Orders

```python
query = """
SELECT *
FROM clean_dataset
"""

pd.read_sql(query, db)
```

# Show Only Delivered Orders

```python
query = """
SELECT OrderID,
       CustomerID,
       Product,
       TotalPrice
FROM clean_dataset
WHERE OrderStatus = 'Delivered'
"""

pd.read_sql(query, db)
```

# Find Orders Above ₹2000

```python
query = """
SELECT OrderID,
       Product,
       TotalPrice
FROM clean_dataset
WHERE TotalPrice > 2000
"""

pd.read_sql(query, db)
```

# Sort Orders by Highest Revenue

```python
query = """
SELECT OrderID,
       Product,
       TotalPrice
FROM clean_dataset
ORDER BY TotalPrice DESC
"""

pd.read_sql(query, db)
```

# Count Total Orders

```python
query = """
SELECT COUNT(*) AS Total_Orders
FROM clean_dataset
"""

pd.read_sql(query, db)
```

# Total Revenue Generated

```python
query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM clean_dataset
"""

pd.read_sql(query, db)
```

# Average Order Value

```python
query = """
SELECT AVG(TotalPrice) AS Average_Order_Value
FROM clean_dataset
"""

pd.read_sql(query, db)
```

# Total Revenue by Product

```python
query = """
SELECT Product,
       SUM(TotalPrice) AS Revenue
FROM clean_dataset
GROUP BY Product
ORDER BY Revenue DESC
"""

pd.read_sql(query, db)
```

# Number of Orders by Payment Method

```python
query = """
SELECT PaymentMethod,
       COUNT(*) AS Total_Orders
FROM clean_dataset
GROUP BY PaymentMethod
ORDER BY Total_Orders DESC
"""

pd.read_sql(query, db)
```

# Orders by Referral Source

```python
query = """
SELECT ReferralSource,
       COUNT(*) AS Orders
FROM clean_dataset
GROUP BY ReferralSource
ORDER BY Orders DESC
"""

pd.read_sql(query, db)
```

# Average Revenue by Order Status

```python
query = """
SELECT OrderStatus,
       AVG(TotalPrice) AS Avg_Revenue
FROM clean_dataset
GROUP BY OrderStatus
ORDER BY Avg_Revenue DESC
"""

pd.read_sql(query, db)
```

# Top 5 Products by Quantity Sold

```python
query = """
SELECT Product,
       SUM(Quantity) AS Total_Quantity
FROM clean_dataset
GROUP BY Product
ORDER BY Total_Quantity DESC
LIMIT 5
"""

pd.read_sql(query, db)
```

# Products Generating Revenue Above Average

```python
query = """
SELECT Product,
       SUM(TotalPrice) AS Revenue
FROM clean_dataset
GROUP BY Product
HAVING Revenue >
(
    SELECT AVG(TotalPrice)
    FROM clean_dataset
)
ORDER BY Revenue DESC
"""

pd.read_sql(query, db)
```

# Top 5 Customers by Spending

```python
query = """
SELECT CustomerID,
       SUM(TotalPrice) AS Total_Spending
FROM clean_dataset
GROUP BY CustomerID
ORDER BY Total_Spending DESC
LIMIT 5
"""

pd.read_sql(query, db)
```

# Revenue by Payment Method

```python
query = """
SELECT PaymentMethod,
       SUM(TotalPrice) AS Revenue
FROM clean_dataset
GROUP BY PaymentMethod
ORDER BY Revenue DESC
"""

pd.read_sql(query, db)
```

# Order Status Distribution

```python
query = """
SELECT OrderStatus,
       COUNT(*) AS Total_Orders
FROM clean_dataset
GROUP BY OrderStatus
ORDER BY Total_Orders DESC
"""

pd.read_sql(query, db)
```

# Orders Using Coupons

```python
query = """
SELECT COUNT(*) AS Orders_With_Coupons
FROM clean_dataset
WHERE CouponCode IS NOT NULL
AND CouponCode <> ''
"""

pd.read_sql(query, db)
```
